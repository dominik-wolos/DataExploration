Imports & paths

```python
import pandas as pd
from pathlib import Path

processed_path = Path("data/processed/cleaned_games.csv")
features_path = Path("data/processed/features.csv")

processed_path, features_path
```

Load data

```python
df_games = pd.read_csv(processed_path)
df_features = pd.read_csv(features_path)

df_games.head(), df_features.head()
```

Basic info & missing values

```python
print("=== Cleaned games info ===")
display(df_games.info())
print("\nMissing values per column:")
display(df_games.isna().sum())

print("\n=== Features info ===")
display(df_features.info())
print("\nMissing values per column:")
display(df_features.isna().sum())
```

Result distribution

```python
print("Game result distribution (raw):")
print(df_games["result"].value_counts(normalize=True))

if "result" in df_features.columns:
    print("\nResult (numeric / parsed) distribution:")
    print(df_features["result"].value_counts(normalize=True))
```

Rating & rating_diff checks

```python
cols = [c for c in df_features.columns if "rating" in c or "elo" in c]
print("Rating-related columns:", cols)

if "rating_diff" in df_features.columns:
    print("\nRating diff summary:")
    display(df_features["rating_diff"].describe())
```

Simple plot

```python
import matplotlib.pyplot as plt

if "rating_diff" in df_features.columns:
    df_features["rating_diff"].hist(bins=50)
    plt.title("Distribution of rating_diff")
    plt.xlabel("rating_diff (white - black)")
    plt.ylabel("count")
    plt.show()
```
